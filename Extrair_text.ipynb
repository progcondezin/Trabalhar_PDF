```python
from PyPDF2 import PdfReader
import PyPDF2


#creating a pdf reader object
reader = PdfReader('KIARA.pdf')
text = ""
for i in range(3):
    page = reader.pages[i]
    # extracting text from page and add it to the text string
    text += page.extract_text()

print(text)
```

```python
pdf = open('KIARA.pdf' , 'rb')
reader = PyPDF2.PdfFileReader(pdf)
pagina = reader.getPage(2)

print(pagina.extract_text())
```
